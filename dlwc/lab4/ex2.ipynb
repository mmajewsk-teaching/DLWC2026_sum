## Level 2: OR Gate Implementation
OR is **linearly separable**: only (0,0) maps to 0, the rest to 1.
A single line can separate these — so a single neuron is enough.

```python
# %%
X = np.array([
    [0, 0, 1],
    [1, 0, 1],
    [0, 1, 1],
    [1, 1, 1]
])
```

```python
Y = np.array([
    [0],
    [1],
    [1],
    [1]
])
```
